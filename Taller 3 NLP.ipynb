### Punto 1
- [4pts] Leer el archivo UNA_SOLITARIA_VOZ_HUMANA.txt con with open
- [4pts] Convertir a minúsculas
- [4pts] Conservar sólo palabras
- [4pts] Tokenizar

```python
import pandas as pd
import os
with open ("C:/Users/practicas/Documents/Especializacion/NLP/Talleres_NLP/Archivos Talleres/UNA_SOLITARIA_VOZ_HUMANA.txt", encoding= "utf-8") as archivo:
    texto = archivo.read()
print(texto)
```

```python
# Convertir en minusculas
import re
minusculas= texto.lower() 
print(minusculas)
```

```python
# Conservar Solo palabras
palabras = re.sub(r"[\W\d_]+"," ", minusculas)
print(palabras)
```

```python
#tokenización
tokenizado = palabras.split()
print(tokenizado)
```

### Punto 2
- [8pts] Quitar palabras vacias
- [9pts] ¿Cuáles son las 10 palabras no vacias más usadas?

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_sp = stopwords.words('spanish')
```

```python
# Quitar palabras vacias
texto = [palabra for palabra in tokenizado if palabra not in stopwords_sp]
texto
```

```python
# las palabras mas usadas
from collections import Counter
Counter(texto).most_common(10)
```

- Tercer Punto
- [8pts] Stemming del documento
- [9pts] ¿Cuáles son las 10 raíces más usadas?

```python
#stemming del documento
from nltk.stem.snowball import SnowballStemmer
spanishStemmer = SnowballStemmer("spanish")
texto_stem = []
for palabra in texto:
    raiz = spanishStemmer.stem(palabra)
    texto_stem.append(raiz)
```

```python
#Raices mas usadas
Counter(texto_stem).most_common(10)
```
