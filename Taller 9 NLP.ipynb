```python
import numpy as np

from wordcloud import WordCloud
from wordcloud import ImageColorGenerator

from nltk.corpus import stopwords
stopwords = stopwords.words('spanish')

import matplotlib.pyplot as plt
import PIL.Image
from IPython.display import display
```

```python
#Cargar Texto

import pandas as pd
data = pd.read_csv("C:/Users/practicas/Desktop/Princesas.csv")
data.head()
```

```python
palabras = data.Personalidad.str.cat(sep=' ')

wordcloud = WordCloud(width=800, height=400,
                     max_font_size=150, max_words=250,
                     background_color='white', colormap='cool', # gist het = color
                     stopwords=stopwords).generate(palabras)
#Dibujar  figura
plt.figure(figsize=(10,8)) #tamaño
plt.imshow(wordcloud, interpolation='bilinear') #la muestra interpolation mejora la calidad de la imagen
plt.axis("off") # apaga los ejes
plt.tight_layout(pad=0) # fondo ajustado
plt.show()
```

```python
palabras = data.Personalidad.str.cat(sep=' ')

 

wordcloud = WordCloud(width=800, height=400,
                     max_font_size=150, max_words=250,
                     background_color='white', colormap='hsv',
                     stopwords=stopwords,
                     mask=mask, contour_width=0.5, contour_color='pink').generate(palabras) # controno de mascara

 
wordcloud.to_file("Princes.png") # calidad del archivo

 

plt.figure(figsize=(10,8)) #tamaño mascara
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
